```python
import pandas as pd
```

```python
#Read csv into a dataframe
vine_df = pd. read_csv("vine_table.csv")
vine_df.head()
```

```python
#Filter the dataframe for rows where total_votes is equal to or greater than 20
total_votes_filter_df = vine_df[vine_df["total_votes"] >= 20]
total_votes_filter_df
```

```python
#Filter the dataframe to create a new dataframe where the number of helpful_votes divided by the total_votes is equal to or greater than 50%
greater_than_50_df = total_votes_filter_df[total_votes_filter_df["helpful_votes"] / total_votes_filter_df["total_votes"] >= 0.5]
greater_than_50_df
```

```python
#Filter the dataframe and create a new dataframe that retrieves all the rows where a review was written as part of the vine program
vine_reviews_df = greater_than_50_df[greater_than_50_df["vine"] == "Y"]
vine_reviews_df
```

```python
#Reviews that are not part of the vine program
unpaid_reviews_df = greater_than_50_df[greater_than_50_df["vine"] == "N"]
unpaid_reviews_df
```

```python
#Total Number of reviews
paid_reviews_total = len(vine_reviews_df)
unpaid_reviews_total = len(unpaid_reviews_df)

#Number of 5-star reviews
paid_5star_reviews = len(vine_reviews_df[vine_reviews_df["star_rating"] == 5.0])
unpaid_5star_reviews = len(unpaid_reviews_df[unpaid_reviews_df["star_rating"] == 5.0])

#Percentage of 5-star reviews for paid vs unpaid
paid_percentage = (paid_5star_reviews/paid_reviews_total)*100
print(paid_percentage)
unpaid_percentage = (unpaid_5star_reviews/unpaid_reviews_total)*100
print(unpaid_percentage)
```
